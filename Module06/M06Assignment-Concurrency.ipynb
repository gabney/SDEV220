### 13.1
Write the current date as a string to the text file today.txt.

```python
from datetime import datetime

now = datetime.now() # creates a datetime object with current date and time attributes

# opens a filestream to today.txt in write+ mode - will erase the file and allow reading and writing
with open("today.txt","w+") as f: 
    f.write(now.isoformat()) # writes the iso formatted string of the datetime object

print(f"Wrote the time '{now.isoformat()}' to the file today.txt.") # confirmation message
```

### 13.2
Read the text file today.txt into the string today_string.

```python
# opens the today.txt file in read mode
with open("today.txt","r") as f: 
    today_string = f.read() # saves the contents of the file to today_string

print(f"Read the time '{today_string}' from the file today.txt.")
```

### 13.3
Parse the date from today_string.

```python
# checks to see if the datetime_object can be generated from data that was read from the file
try:
    # uses the 'fromisoformat' method from datetime to convert the iso-formatted string into a new datetime object
    datetime_object = datetime.fromisoformat(today_string)
    print(f"New datetime object generated successfully with value {datetime_object}")
except:
    print("Datetime object could not be generated from file.")
```

### 15.1
Use multiprocessing to create three separate processes. Make each one wait a random number of seconds between zero and one, print the current time, and then exit.

```python
"""
Note: This program may not record the processes in a jupyter notebook. \
Try running the file 'M06_chapter_15_test.py' with identical code to this \
cell to generate the outputs to console.
"""

import multiprocessing
import random
import time


def rand_wait(i):
    """
    Waits a random amount of time between 0 and 1 seconds, then returns the current time.
    """
    # generates a random wait time
    wait_time = random.random()
    print(f"Process {i} will wait for {wait_time} seconds.")
    # waits
    time.sleep(wait_time)
    # saves current time
    now = time.time()
    print(f"Time's up! Process {i}'s current time: {now}")

    


# runs only if the program is the main program
if __name__ == "__main__":
    # creates three processes to run the rand_wait function
    for i in range(3): 
        p = multiprocessing.Process(target = rand_wait, args = (i+1,))
        p.start()
```
