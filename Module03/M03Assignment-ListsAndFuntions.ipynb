### 9.2
Define a generator function called get_odds() that returns the odd numbers from range(10). Use a for loop to find and print the third value returned.

```python
def get_odds(odds_range = 10): #default range of 10
    for odd in range(odds_range):
        if odd % 2 != 0:
            yield odd

odds_tracker = get_odds(10) #creates the generator object
iter_tracker: int = 0 #used to track the number of times the generator has been iterated

for i in odds_tracker:
    iter_tracker += 1
    if iter_tracker == 3:
        print(i)
```
