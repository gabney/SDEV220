## OOP Programming
Prompt from the tutorial at [GeeksForGeeks - Sort 0s 1s and 2s](https://practice.geeksforgeeks.org/problems/sort-an-array-of-0s-1s-and-2s4231/1):

Given an array arr[] containing only 0s, 1s, and 2s. Sort the array in ascending order.\
Note: You need to solve this problem without utilizing the built-in sort function.

```python
def qsorter(arr):
    """
    Function takes an input array and sorts it into left,
    right and middle then returns those parts.  This is a 
    version of the qsort that potentnially uses recursion
    to break down lists into three parts that can be 
    further divided until sorted.
    """
    if len(arr) <= 1: # checks to see if array is empty or single value
        return arr # ends sort
    
    pivot = arr[len(arr) // 2] # sets the pivot to halway through input array
    
    left: list = [x for x in arr if x < pivot] # places all values in arr that are lower than the pivot value in new array called left 
    mid: list = [x for x in arr if x == pivot] # places all values in arr that are equal to the pivot value in new array called mid
    right: list = [x for x in arr if x > pivot] # places all values in arr that are lower than the pivot value in new array called left 

    return qsorter(left) + mid + qsorter(right) # returns three parts of list divided around pivot. left and right values are recursively passed to function to check if they need to be sorted further


test_arr: list = [0,1,2,2,1,0,0,1,2] # test array to attempt to sort

test_arr_sorted: list = qsorter(test_arr) # runs qsorter on test array and saves as new array

print(f"The sorted list is {test_arr_sorted}")
```

## Functional Programming
Prompt from the tutorial at [GeeksForGeeks - Binary Search](https://www.geeksforgeeks.org/problems/binary-search-1587115620/1):

Given a sorted array arr[] and an integer k, find the position(0-based indexing) at which k is present in the array using binary search. If k doesn't exist in arr[] return -1.\
Note: If multiple occurrences are there, please return the smallest index.

```python
def binary_search(unsearched_list, search_value): # creates search function with list to search and value to search for
    """
    This function accepts an organized list and a search value, and 
    searches for the value in that list using binary search technique.
    """

    low = 0 # lowest index value of list to search starting with 0
    high = len(unsearched_list) - 1 # highest index value of list to search set to the last value in the list's length
    
    while low <= high: # loop to set mid value, check mid value against the search value, and adjust list search 
        
        mid = low + (high - low) // 2 # sets mid to the value of the midpoint of the input list
        
        if unsearched_list[mid] == search_value: #checks to see if the mid value is the search value
            return mid # returns the search value if found, breaking out of the function
        
        elif unsearched_list[mid] < search_value: # if midpoint is less than search value, moves the lower part of the search to be mid + 1, the repeat the while loop
            low = mid + 1

        else: # if midpoint is greater than search value, moves the upper part of the search to be mid - 1, then repeat the while loop
            high = mid - 1

    return -1 # returns -1 value if the search value is not found in the list 

# uses the function with a test list and integer value to search for in that list
test_list: list = [-999, -99, -9, -1, 0, 1, 9, 42, 99, 999, 9999]
test_value: int = 42

# performs the binary search for the test value on the test list
test_result = binary_search(test_list, test_value)

print(f"{test_value} is found in the list at index {test_result}.")
```
